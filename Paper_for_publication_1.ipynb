```python

# TIFA: Three-Phase Inflation and Field Alignment
## A Single-Field Model of Inflation and Dark Energy
### Consistent with Planck 2018 and DESI DR1

**Author:** Tony (Brother Tony)
**Repository:** github.com/[your-handle]/tifa-cosmology
**Date:** March 2026
**Status:** Preprint v1.0 — open for community review
**License:** MIT

---

> *"The most we can claim is consistency.
>  The most we can offer is falsifiability.
>  The rest belongs to the experiments."*

---

## Abstract

We present TIFA (Three-phase Inflation and Field
Alignment), a single-field cosmological model in which
one canonical scalar field on a cosine potential drives
both primordial inflation and the present acceleration
of the universe. The inflation sector, operating at
decay constant f = 7 M_Pl, predicts a spectral index
n_s = 0.969 and tensor-to-scalar ratio r = 0.041,
consistent with Planck 2018 and testable by LiteBIRD
in 2028. The dark energy sector, operating at effective
decay constant f_eff = 0.5 M_Pl, reproduces the
correct dark energy density today and fits the DESI DR1
data with chi2 = 2.014, compared to chi2 = 11.052 for
Lambda-CDM, a Delta chi2 = 9.038 improvement with two
additional parameters (Delta AIC = 5.038). The full
w(z) trajectory departs from the CPL linearisation by
|Delta w| = 0.13 at z = 3, exceeding Euclid forecast
precision by a factor of five. The model predicts no
phantom crossing at any redshift. Three falsifiable
predictions are identified with specific experimental
targets in 2025, 2026, and 2028. Open theoretical
questions — the derivation of f_eff from reheating and
the cosmological constant problem — are acknowledged
explicitly.

---

## 1. Introduction

Two of the deepest problems in modern cosmology are the
origin of primordial inflation and the nature of dark
energy. They are almost always treated separately.
Inflation is assigned to a high-energy inflaton field
active in the very early universe. Dark energy is
assigned to a cosmological constant, or occasionally to
a separate quintessence field active today.

This separation is theoretically unsatisfying. It
doubles the number of unexplained scalar fields in
cosmology without physical motivation. The question
arises naturally: can a single scalar field, with a
single potential, account for both epochs?

The cosine potential — natural in the context of
axion-like fields and natural inflation — is a
candidate. It possesses a flat hilltop region suitable
for slow-roll inflation at large field values, and a
shallow rolling region suitable for dark energy at
small field values. The two regimes are controlled by
different effective parameters, reflecting the
possibility that the field's effective coupling changes
between epochs, as occurs in models with field
alignment or kinetic mixing.

In this paper we develop TIFA (Three-phase Inflation
and Field Alignment), a concrete single-field
implementation of this idea. We compute explicit
predictions for inflation observables, fit the dark
energy sector to DESI DR1 data, and compute the full
w(z) trajectory from first principles. We identify
three specific falsifiable predictions with target
experiments and dates.

We do not claim to have solved the cosmological
constant problem. We do not claim to have derived
f_eff from microphysics. We claim internal
consistency, agreement with current data, and
falsifiability by near-future experiments. That is
what this paper demonstrates.

### 1.1 Timing and motivation

The DESI DR1 release (2024) reported a preference for
dynamical dark energy at the 2-3 sigma level, with
w0 = -0.827 ± 0.059 and wa = -0.750 ± 0.290 under
the CPL parameterisation. Lambda-CDM, with w = -1
fixed, lies outside the 2-sigma contour of this
measurement. This creates a window for dynamical dark
energy models to be tested seriously.

TIFA was developed in this context. Its predictions
were fixed before DESI DR2 and before Euclid data.
They are therefore genuine predictions, not fits to
future data.

---

## 2. The Model

### 2.1 The potential

TIFA uses a single cosine potential of the form
```

```python


Three features of this trajectory are physically
significant:

**Feature 1 — No phantom crossing.**
w(z) > -1 at all redshifts. The field rolls on a
canonical kinetic term. No ghost instability is
present. No Big Rip occurs.

**Feature 2 — CPL breakdown at z > 0.7.**
The CPL parameterisation, used by DESI to report
their results, predicts w < -1 at z > 1 when fitted
to TIFA's w0 and wa values. This is unphysical for
a canonical scalar field. The TIFA trajectory remains
physical throughout. CPL is an approximation that
breaks down at high redshift.

**Feature 3 — Detectable curvature.**
The accumulated departure between TIFA and CPL
reaches |Delta w| = 0.13 at z = 3. Euclid's
forecast precision is sigma_w ~ 0.025 per redshift
bin. The TIFA signature is therefore detectable at
approximately 5 sigma by Euclid if the model is
correct.

The quantitative falsification statement is:

> *The TIFA equation of state departs from the CPL
> parameterisation by |Delta w| = 0.13 at z = 3,
> exceeding the Euclid 1-sigma precision band of
> 0.025 by a factor of five. The departure begins
> near z = 0.7, grows monotonically with redshift,
> and remains above w = -1 throughout. If Euclid
> measures w(z) consistent with CPL at z > 1 and
> inconsistent with TIFA at greater than 2 sigma,
> the model is falsified.*

---

## 5. Discussion

### 5.1 What TIFA claims

TIFA claims three things:

1. A single cosine potential can produce inflation
   consistent with Planck 2018 at f = 7 M_Pl.

2. The same potential, at f_eff = 0.5 M_Pl,
   produces dark energy consistent with DESI DR1
   with chi2 = 2.014.

3. The full w(z) trajectory is a unique, detectable,
   falsifiable prediction that distinguishes TIFA
   from Lambda-CDM and from CPL.

### 5.2 What TIFA does not claim

TIFA does not claim:

1. To solve the cosmological constant problem.
   Lambda is fitted, not derived. The question of
   why the vacuum energy is not 10^120 times larger
   remains open.

2. To derive f_eff from reheating. The change in
   effective decay constant between epochs is assumed,
   not derived. Deriving it requires a specific
   reheating model and is the subject of Paper 2.

3. To be the unique model consistent with the data.
   Other quintessence models also fit DESI DR1.
   TIFA's advantage is the unified single-field
   structure and the specific falsifiable predictions.

4. To have passed peer review. This is a GitHub
   preprint. It is open for community scrutiny.

### 5.3 Open questions

The following questions are identified for future
work:

**Q1 (reheating):** What microphysical mechanism
produces the change from f = 7 M_Pl to
f_eff = 0.5 M_Pl? This requires a specific
reheating model with field alignment or kinetic
mixing. This is Paper 2.

**Q2 (stability):** The background evolution is
stable (w > -1, canonical kinetic term). A full
perturbation theory analysis — computing the sound
speed cs^2 and checking for gradient instabilities
— is an important cross-check. For canonical
quintessence cs^2 = 1 generically, but this should
be verified explicitly. This is Appendix A of
Paper 1.5.

**Q3 (joint fit):** We have fitted w0 and wa to
DESI DR1 only. A joint fit to Planck CMB power
spectrum, BAO (DESI DR1), and Pantheon+ supernovae
is needed for a complete phenomenological paper.
This is Paper 1.5.

**Q4 (H0 tension):** Does TIFA's w(z) evolution
help or hurt the H0 tension? This is an open
calculation.

### 5.4 The significance of the w(z) table

The w(z) table in Section 4.3 is the most important
output of this paper. It is important for a reason
that goes beyond the fit quality.

At z > 0.7, CPL — the parameterisation used by DESI
to report their results — predicts phantom behaviour
(w < -1). TIFA predicts non-phantom behaviour
(w > -1). These are qualitatively different
physical predictions about the same universe.
Euclid will measure w(z) in this redshift range.
The measurement will distinguish them.

This means the DESI DR1 result, if interpreted
through CPL, may be pointing toward phantom dark
energy. But if the underlying physics is TIFA,
the correct interpretation is non-phantom
quintessence whose CPL approximation happens to
look phantom in the (w0, wa) plane. This is a
physical distinction with observable consequences.

---

## 6. Conclusion

We have presented TIFA, a single-field cosmological
model unifying inflation and dark energy on a cosine
potential. The model produces:
```
