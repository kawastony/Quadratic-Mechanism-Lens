```python

import numpy as np
from scipy.integrate import solve_ivp, quad
from scipy.optimize import minimize_scalar, minimize
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("TIFA: WHAT f DOES DESI WANT?")
print("Free parameter scan over f")
print("="*60)

Omega_m  = 0.315
Omega_DE = 0.685
Lambda4  = 3.0 * Omega_DE
N_start  = np.log(1e-4)
c_km     = 299792.458
rd       = 147.09

f_TIFA   = 3.0/np.sqrt(97.0)
print(f"\n  Λ⁴     = {Lambda4:.6f}")
print(f"  f_TIFA = {f_TIFA:.6f} "
      f"[original]")
print()

DESI_DATA = [
    (0.30, 'DV_rd',  7.93,  0.15),
    (0.51, 'DM_rd', 13.62,  0.25),
    (0.51, 'DH_rd', 20.98,  0.61),
    (0.71, 'DM_rd', 16.85,  0.32),
    (0.71, 'DH_rd', 20.08,  0.60),
    (0.93, 'DM_rd', 21.71,  0.28),
    (0.93, 'DH_rd', 17.88,  0.35),
    (1.32, 'DM_rd', 27.79,  0.69),
    (1.32, 'DH_rd', 13.82,  0.42),
    (1.49, 'DM_rd', 30.21,  0.79),
    (1.49, 'DH_rd', 13.23,  0.55),
    (2.33, 'DM_rd', 39.71,  0.94),
    (2.33, 'DH_rd',  8.52,  0.17),
]
N_data = len(DESI_DATA)

##────────────────────────────────────────
## SOLVER — f as parameter
##────────────────────────────────────────

def solve_tifa_f(phi_i, f):
    def ode(N, y):
        phi, dphi = y
        a     = np.exp(N)
        rho_m = Omega_m * a**(-3)
        V     = Lambda4*(1.0+np.cos(phi/f))
        A     = dphi**2
        denom = max(3.0 - A/2.0, 0.01)
        H2    = max((rho_m+V)/denom, 1e-15)
        for _ in range(20):
            Vn = Lambda4*(1.0+np.cos(phi/f))
            F  = (3.0-A/2.0)*H2 - rho_m - Vn
            H2 = max(
                H2 - F/max(3.0-A/2.0,1e-15),
                1e-15
            )
        Vp    = -(Lambda4/f)*np.sin(phi/f)
        eps_H = min(
            (rho_m + H2*A)/(2.0*H2), 3.0
        )
        ddphi = -(3.0-eps_H)*dphi - Vp/H2
        return [dphi, ddphi]

    return solve_ivp(
        ode, [N_start, 0.0],
        [phi_i, 0.0],
        method='DOP853',
        max_step=0.002,
        rtol=1e-11, atol=1e-13,
        dense_output=True
    )

def get_state_f(sol, N, f):
    y         = sol.sol(N)
    phi, dphi = y[0], y[1]
    a         = np.exp(N)
    rho_m     = Omega_m * a**(-3)
    V         = Lambda4*(1.0+np.cos(phi/f))
    A         = dphi**2
    H2        = max(
        (rho_m+V)/max(3.0-A/2.0,0.01),
        1e-15
    )
    for _ in range(20):
        Vn = Lambda4*(1.0+np.cos(phi/f))
        F  = (3.0-A/2.0)*H2 - rho_m - Vn
        H2 = max(
            H2 - F/max(3.0-A/2.0,1e-15),
            1e-15
        )
    KE       = H2*A/2.0
    rho_phi  = KE + V
    Omega_p  = rho_phi/(3.0*H2)
    w_phi    = (KE-V)/max(rho_phi, 1e-15)
    return {
        'H2':H2, 'V':V, 'KE':KE,
        'phi':phi, 'dphi':dphi,
        'Omega_phi':Omega_p,
        'w_phi':w_phi, 'rho_m':rho_m
    }

##────────────────────────────────────────
## CHI2 FUNCTION: optimize over φ_i, H0
##────────────────────────────────────────

def chi2_full(f, phi_i_guess=None):
    """
    For given f:
    1. Find φ_i s.t. Ω_φ(0)=0.685
    2. Optimize H0
    Returns (chi2, H0, w0, phi_i, sol)
    """
    from scipy.optimize import brentq

    # Step 1: bracket φ_i for Ω_φ = 0.685
    def residual(phi_i):
        sol = solve_tifa_f(phi_i, f)
        if not sol.success:
            return 1.0
        st = get_state_f(sol, 0.0, f)
        return st['Omega_phi'] - Omega_DE

    # Scan for bracket
    phi_test = np.linspace(0.001, np.pi*f*0.99,
                            60)
    res_arr  = []
    for pt in phi_test:
        try:
            r = residual(pt)
        except Exception:
            r = np.nan
        res_arr.append(r)

    # Find sign change
    bracket = None
    for i in range(len(phi_test)-1):
        if (not np.isnan(res_arr[i]) and
            not np.isnan(res_arr[i+1])):
            if res_arr[i]*res_arr[i+1] < 0:
                bracket = (phi_test[i],
                           phi_test[i+1])
                break

    if bracket is None:
        return 1e10, 70.0, -1.0, np.nan, None

    try:
        phi_i_opt = brentq(
            residual, bracket[0], bracket[1],
            xtol=1e-8
        )
    except Exception:
        return 1e10, 70.0, -1.0, np.nan, None

    sol = solve_tifa_f(phi_i_opt, f)
    if not sol.success:
        return 1e10, 70.0, -1.0, np.nan, None

    st0 = get_state_f(sol, 0.0, f)
    H2_0 = st0['H2']
    w0   = st0['w_phi']

    # Step 2: optimize H0
    def E2(z):
        N  = max(np.log(1.0/(1.0+z)), N_start)
        st = get_state_f(sol, N, f)
        return st['H2'] / H2_0

    def chi2_H0(H0):
        total = 0.0
        for (z, typ, val, sig) in DESI_DATA:
            if typ == 'DM_rd':
                ig, _ = quad(
                    lambda zp: 1.0/np.sqrt(
                        max(E2(zp),1e-10)
                    ),
                    0, z, limit=80,
                    epsabs=1e-8, epsrel=1e-8
                )
                pred = (c_km/H0)*ig/rd
            elif typ == 'DH_rd':
                pred = (c_km/H0)/(
                    np.sqrt(max(E2(z),1e-10))*rd
                )
            elif typ == 'DV_rd':
                ig, _ = quad(
                    lambda zp: 1.0/np.sqrt(
                        max(E2(zp),1e-10)
                    ),
                    0, z, limit=80,
                    epsabs=1e-8, epsrel=1e-8
                )
                DM = (c_km/H0)*ig/rd
                DH = (c_km/H0)/(
                    np.sqrt(max(E2(z),1e-10))*rd
                )
                pred = (z*DM**2*DH)**(1.0/3.0)
            total += ((pred-val)/sig)**2
        return total

    res_H = minimize_scalar(
        chi2_H0,
        bounds=(60.0, 90.0),
        method='bounded',
        options={'xatol':1e-5}
    )
    return res_H.fun, res_H.x, w0, phi_i_opt, sol

##────────────────────────────────────────
## SCAN OVER f
##────────────────────────────────────────

print("─"*70)
print("SCANNING f: does larger f improve χ²?")
print("─"*70)
print()
print(f"  {'f':>8}  "
      f"{'f/f_T':>7}  "
      f"{'πf':>6}  "
      f"{'φ_i':>8}  "
      f"{'φ_i/f':>7}  "
      f"{'w(0)':>8}  "
      f"{'H0':>7}  "
      f"{'χ²':>10}")
print("  " + "─"*68)

# Scan f from 0.1 to 5.0 M_Pl
f_arr  = np.concatenate([
    np.linspace(0.10, 0.50, 10),
    np.linspace(0.60, 2.00, 15),
    np.linspace(2.50, 5.00,  6),
])

results = []
for f in f_arr:
    c2, H0, w0, phi_i, sol = chi2_full(f)
    results.append((f, c2, H0, w0, phi_i, sol))

    flag = " ←" if c2 < 50 else ""
    print(f"  {f:8.4f}  "
          f"{f/f_TIFA:7.3f}  "
          f"{np.pi*f:6.3f}  "
          f"{phi_i:8.5f}  "
          f"{phi_i/f if f>0 else 0:7.4f}  "
          f"{w0:8.4f}  "
          f"{H0:7.3f}  "
          f"{c2:10.3f}{flag}")

print()

##────────────────────────────────────────
## ΛCDM BASELINE
##────────────────────────────────────────

def chi2_LCDM(params):
    H0, Om = params
    total  = 0.0
    for (z, typ, val, sig) in DESI_DATA:
        E2L = lambda zp: Om*(1+zp)**3+(1-Om)
        if typ == 'DM_rd':
            ig, _ = quad(
                lambda zp: 1.0/np.sqrt(
                    max(E2L(zp),1e-10)
                ),
                0, z, limit=80,
                epsabs=1e-8, epsrel=1e-8
            )
            pred = (c_km/H0)*ig/rd
        elif typ == 'DH_rd':
            pred = (c_km/H0)/(
                np.sqrt(max(E2L(z),1e-10))*rd
            )
        elif typ == 'DV_rd':
            ig, _ = quad(
                lambda zp: 1.0/np.sqrt(
                    max(E2L(zp),1e-10)
                ),
                0, z, limit=80,
                epsabs=1e-8, epsrel=1e-8
            )
            DM = (c_km/H0)*ig/rd
            DH = (c_km/H0)/(
                np.sqrt(max(E2L(z),1e-10))*rd
            )
            pred = (z*DM**2*DH)**(1.0/3.0)
        total += ((pred-val)/sig)**2
    return total

res_L  = minimize(
    chi2_LCDM, x0=[67.4, 0.315],
    method='Nelder-Mead',
    options={'xatol':1e-7,'fatol':1e-7,
             'maxiter':50000}
)
chi2_L = res_L.fun
H0_L   = res_L.x[0]
Om_L   = res_L.x[1]

print(f"  ΛCDM: χ²={chi2_L:.3f}  "
      f"H0={H0_L:.2f}  Ωm={Om_L:.4f}")
print()

##────────────────────────────────────────
## FIND BEST f
##────────────────────────────────────────

valid = [(f,c2,H0,w0,phi_i,sol)
         for (f,c2,H0,w0,phi_i,sol)
         in results if c2 < 1e9]

if valid:
    best = min(valid, key=lambda x: x[1])
    f_best, c2_best, H0_best, \
    w0_best, phi_best, sol_best = best

    print("─"*70)
    print("BEST f FOUND")
    print("─"*70)
    print(f"""
  f_best    = {f_best:.6f} M_Pl
  f/f_TIFA  = {f_best/f_TIFA:.4f}
  πf        = {np.pi*f_best:.4f}
  φ_i       = {phi_best:.6f}
  φ_i/f     = {phi_best/f_best:.6f}
  w(z=0)    = {w0_best:.6f}
  H0        = {H0_best:.4f} km/s/Mpc
  χ²_TIFA   = {c2_best:.4f}
  χ²_ΛCDM   = {chi2_L:.4f}
  Δχ²       = {c2_best-chi2_L:+.4f}
""")

##────────────────────────────────────────
## PLOT: χ²(f)
##────────────────────────────────────────

f_pl  = [r[0] for r in valid]
c2_pl = [r[1] for r in valid]
w_pl  = [r[3] for r in valid]
H0_pl = [r[2] for r in valid]

fig, axes = plt.subplots(1, 3, figsize=(15,5))
fig.suptitle(
    'TIFA: χ²(f) scan with Ω_φ(0)=Ω_DE',
    fontsize=13
)

ax = axes[0]
ax.semilogy(f_pl, c2_pl, 'b-o',
            lw=2, ms=5)
ax.axhline(chi2_L, color='k', ls=':',
           lw=1.5,
           label=f'ΛCDM χ²={chi2_L:.1f}')
ax.axvline(f_TIFA, color='r', ls='--',
           lw=1.5,
           label=f'f_TIFA={f_TIFA:.3f}')
if valid:
    ax.axvline(f_best, color='g', ls='--',
               lw=1.5,
               label=f'f_best={f_best:.3f}')
ax.set_xlabel('f [M_Pl]', fontsize=11)
ax.set_ylabel('χ²', fontsize=11)
ax.set_title('χ²(f)', fontsize=11)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

ax = axes[1]
ax.plot(f_pl, w_pl, 'r-o', lw=2, ms=5)
ax.axhline(-1.0, color='k', ls=':',
           label='w=-1')
ax.axhline(-0.7, color='gray', ls='--',
           label='w=-0.7')
ax.axvline(f_TIFA, color='r', ls='--',
           lw=1.5)
ax.set_xlabel('f [M_Pl]', fontsize=11)
ax.set_ylabel('w(z=0)', fontsize=11)
ax.set_title('w(0) vs f', fontsize=11)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

ax = axes[2]
ax.plot(f_pl, H0_pl, 'g-o', lw=2, ms=5)
ax.axhline(67.4, color='k', ls=':',
           label='H0=67.4')
ax.axhline(H0_L, color='b', ls='--',
           label=f'ΛCDM H0={H0_L:.1f}')
ax.axvline(f_TIFA, color='r', ls='--',
           lw=1.5,
           label=f'f_TIFA')
ax.set_xlabel('f [M_Pl]', fontsize=11)
ax.set_ylabel('H0 [km/s/Mpc]', fontsize=11)
ax.set_title('H0(f)', fontsize=11)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('tifa_f_scan.png',
            dpi=150, bbox_inches='tight')
print()
print("  Saved: tifa_f_scan.png")

print()
print("="*70)
print("CONCLUSION")
print("="*70)
print(f"""
  f_TIFA = {f_TIFA:.4f} gives χ² = 830 (terrible)

  The scan will show what f DESI actually wants.

  KEY QUESTION for Brother Tony:
  Is f = 3/√97 FIXED by the theory,
  or is it DERIVED from other parameters?

  If fixed: TIFA cannot fit DESI BAO
  as currently formulated.

  If free: the best-fit f will tell us
  what the theory needs.
""")
```

```python

import numpy as np
from scipy.integrate import solve_ivp, quad
from scipy.optimize import brentq, minimize_scalar
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("TIFA: FINE SCAN NEAR f_best = 2.0")
print("="*60)

Omega_m  = 0.315
Omega_DE = 0.685
Lambda4  = 3.0 * Omega_DE
N_start  = np.log(1e-4)
c_km     = 299792.458
rd       = 147.09
f_TIFA   = 3.0/np.sqrt(97.0)

DESI_DATA = [
    (0.30, 'DV_rd',  7.93,  0.15),
    (0.51, 'DM_rd', 13.62,  0.25),
    (0.51, 'DH_rd', 20.98,  0.61),
    (0.71, 'DM_rd', 16.85,  0.32),
    (0.71, 'DH_rd', 20.08,  0.60),
    (0.93, 'DM_rd', 21.71,  0.28),
    (0.93, 'DH_rd', 17.88,  0.35),
    (1.32, 'DM_rd', 27.79,  0.69),
    (1.32, 'DH_rd', 13.82,  0.42),
    (1.49, 'DM_rd', 30.21,  0.79),
    (1.49, 'DH_rd', 13.23,  0.55),
    (2.33, 'DM_rd', 39.71,  0.94),
    (2.33, 'DH_rd',  8.52,  0.17),
]
N_data = len(DESI_DATA)

##────────────────────────────────────────
## SOLVER
##────────────────────────────────────────

def solve_tifa_f(phi_i, f):
    def ode(N, y):
        phi, dphi = y
        a     = np.exp(N)
        rho_m = Omega_m * a**(-3)
        V     = Lambda4*(1.0+np.cos(phi/f))
        A     = dphi**2
        H2    = max(
            (rho_m+V)/max(3.0-A/2.0,0.01),
            1e-15
        )
        for _ in range(20):
            Vn = Lambda4*(1.0+np.cos(phi/f))
            F  = (3.0-A/2.0)*H2-rho_m-Vn
            H2 = max(
                H2-F/max(3.0-A/2.0,1e-15),
                1e-15
            )
        Vp    = -(Lambda4/f)*np.sin(phi/f)
        eps_H = min(
            (rho_m+H2*A)/(2.0*H2), 3.0
        )
        ddphi = -(3.0-eps_H)*dphi - Vp/H2
        return [dphi, ddphi]

    return solve_ivp(
        ode, [N_start, 0.0],
        [phi_i, 0.0],
        method='DOP853',
        max_step=0.001,
        rtol=1e-12, atol=1e-14,
        dense_output=True
    )

def get_state(sol, N, f):
    y         = sol.sol(N)
    phi, dphi = y[0], y[1]
    a         = np.exp(N)
    rho_m     = Omega_m * a**(-3)
    V         = Lambda4*(1.0+np.cos(phi/f))
    A         = dphi**2
    H2        = max(
        (rho_m+V)/max(3.0-A/2.0,0.01),
        1e-15
    )
    for _ in range(20):
        Vn = Lambda4*(1.0+np.cos(phi/f))
        F  = (3.0-A/2.0)*H2-rho_m-Vn
        H2 = max(
            H2-F/max(3.0-A/2.0,1e-15),
            1e-15
        )
    KE      = H2*A/2.0
    rho_phi = KE+V
    return {
        'H2':H2, 'V':V, 'KE':KE,
        'phi':phi, 'dphi':dphi,
        'Omega_phi':rho_phi/(3.0*H2),
        'w_phi':(KE-V)/max(rho_phi,1e-15),
        'rho_m':rho_m
    }

def get_phi_i(f):
    """Find phi_i s.t. Omega_phi(0)=0.685"""
    def residual(phi_i):
        sol = solve_tifa_f(phi_i, f)
        if not sol.success:
            return 1.0
        return (get_state(sol,0.0,f)
                ['Omega_phi'] - Omega_DE)

    pts = np.linspace(0.001, np.pi*f*0.999,
                       80)
    rvs = []
    for p in pts:
        try:
            rvs.append((p, residual(p)))
        except Exception:
            pass

    for i in range(len(rvs)-1):
        if rvs[i][1]*rvs[i+1][1] < 0:
            return brentq(
                residual,
                rvs[i][0], rvs[i+1][0],
                xtol=1e-10
            )
    return np.nan

def chi2_tifa(f):
    phi_i = get_phi_i(f)
    if np.isnan(phi_i):
        return 1e10, 70., -1., phi_i

    sol   = solve_tifa_f(phi_i, f)
    if not sol.success:
        return 1e10, 70., -1., phi_i

    H2_0 = get_state(sol, 0.0, f)['H2']
    w0   = get_state(sol, 0.0, f)['w_phi']

    def E2(z):
        N = max(np.log(1/(1+z)), N_start)
        return get_state(sol,N,f)['H2']/H2_0

    def chi2_H0(H0):
        t = 0.0
        for (z,typ,val,sig) in DESI_DATA:
            if typ == 'DM_rd':
                ig,_ = quad(
                    lambda zp:1/np.sqrt(
                        max(E2(zp),1e-10)),
                    0,z,limit=80,
                    epsabs=1e-8,epsrel=1e-8)
                p = (c_km/H0)*ig/rd
            elif typ == 'DH_rd':
                p = (c_km/H0)/(
                    np.sqrt(max(E2(z),1e-10))
                    *rd)
            elif typ == 'DV_rd':
                ig,_ = quad(
                    lambda zp:1/np.sqrt(
                        max(E2(zp),1e-10)),
                    0,z,limit=80,
                    epsabs=1e-8,epsrel=1e-8)
                DM = (c_km/H0)*ig/rd
                DH = (c_km/H0)/(
                    np.sqrt(max(E2(z),1e-10))
                    *rd)
                p = (z*DM**2*DH)**(1/3)
            t += ((p-val)/sig)**2
        return t

    res = minimize_scalar(
        chi2_H0, bounds=(60.,90.),
        method='bounded',
        options={'xatol':1e-6})
    return res.fun, res.x, w0, phi_i

##────────────────────────────────────────
## FINE SCAN f = 1.5 to 3.0
##────────────────────────────────────────

print()
print("─"*65)
print("Fine scan: f = 1.5 to 3.0 M_Pl")
print("─"*65)
print()
print(f"  {'f':>7}  {'φ_i':>8}  "
      f"{'φ_i/f':>7}  {'φ_i/(πf)':>9}  "
      f"{'w(0)':>8}  {'H0':>7}  {'χ²':>8}")
print("  " + "─"*60)

f_fine = np.linspace(1.5, 3.0, 31)
results_fine = []

for f in f_fine:
    c2, H0, w0, phi_i = chi2_tifa(f)
    results_fine.append(
        (f, c2, H0, w0, phi_i)
    )
    ratio = phi_i/f if not np.isnan(phi_i) \
            else np.nan
    frac  = phi_i/(np.pi*f) \
            if not np.isnan(phi_i) else np.nan
    flag  = " ←" if c2 < 16.5 else ""
    print(f"  {f:7.4f}  {phi_i:8.5f}  "
          f"{ratio:7.5f}  {frac:9.6f}  "
          f"{w0:8.5f}  {H0:7.4f}  "
          f"{c2:8.4f}{flag}")

print()

# Find minimum
best = min(results_fine, key=lambda x: x[1])
f_opt, c2_opt, H0_opt, w0_opt, phi_opt = best

print(f"  χ² minimum:")
print(f"  f_opt     = {f_opt:.6f} M_Pl")
print(f"  φ_i       = {phi_opt:.7f}")
print(f"  φ_i/f     = {phi_opt/f_opt:.7f}")
print(f"  φ_i/(πf)  = "
      f"{phi_opt/(np.pi*f_opt):.7f}")
print(f"  w(0)      = {w0_opt:.6f}")
print(f"  H0        = {H0_opt:.4f}")
print(f"  χ²        = {c2_opt:.4f}")
print()

##────────────────────────────────────────
## FULL EVOLUTION AT f_opt
##────────────────────────────────────────

phi_i_opt = get_phi_i(f_opt)
sol_opt   = solve_tifa_f(phi_i_opt, f_opt)
st0       = get_state(sol_opt, 0.0, f_opt)

print("─"*65)
print(f"Evolution at f_opt = {f_opt:.4f}")
print("─"*65)
print()
print(f"  {'z':>6}  {'φ':>9}  "
      f"{'φ/(πf)':>8}  {'Ω_φ':>8}  "
      f"{'Ω_m':>8}  {'w_φ':>8}")
print("  " + "─"*54)

z_tab = [100,50,20,10,5,3,2.33,
         2,1.5,1,0.7,0.5,0.3,0.1,0.0]
for z in z_tab:
    N  = max(np.log(1/(1+z)), N_start)
    st = get_state(sol_opt, N, f_opt)
    Om_m = st['rho_m']/(3*st['H2'])
    frac = st['phi']/(np.pi*f_opt)
    print(f"  {z:6.2f}  "
          f"{st['phi']:9.6f}  "
          f"{frac:8.6f}  "
          f"{st['Omega_phi']:8.5f}  "
          f"{Om_m:8.5f}  "
          f"{st['w_phi']:8.5f}")

print()

##────────────────────────────────────────
## RATIO ANALYSIS
##────────────────────────────────────────

print("─"*65)
print("RATIO ANALYSIS: what is f_opt?")
print("─"*65)
print()
print(f"  f_opt          = {f_opt:.6f}")
print(f"  f_TIFA         = {f_TIFA:.6f}")
print(f"  ratio          = "
      f"{f_opt/f_TIFA:.6f}")
print(f"  f_opt²         = "
      f"{f_opt**2:.6f}")
print(f"  1/f_opt²       = "
      f"{1/f_opt**2:.6f}")
print(f"  f_opt/π        = "
      f"{f_opt/np.pi:.6f}")
print(f"  f_opt×√97      = "
      f"{f_opt*np.sqrt(97):.6f}")
print(f"  f_opt×√(Ω_m)   = "
      f"{f_opt*np.sqrt(Omega_m):.6f}")
print(f"  f_opt²×97      = "
      f"{f_opt**2*97:.6f}")
print()

# Key cosmological ratios
print(f"  Ω_DE/Ω_m       = "
      f"{Omega_DE/Omega_m:.6f}")
print(f"  √(Ω_DE/Ω_m)    = "
      f"{np.sqrt(Omega_DE/Omega_m):.6f}")
print(f"  (Ω_DE/Ω_m)^(1/3)= "
      f"{(Omega_DE/Omega_m)**(1/3):.6f}")
print(f"  Λ⁴             = {Lambda4:.6f}")
print(f"  √Λ⁴            = "
      f"{np.sqrt(Lambda4):.6f}")
print(f"  Λ⁴/f_opt²      = "
      f"{Lambda4/f_opt**2:.6f}")
print()

##────────────────────────────────────────
## FINAL PLOTS
##────────────────────────────────────────

z_plot = np.linspace(0, 3, 400)
w_ev, Op_ev, phi_ev, E_ev = [],[],[],[]
H2_0 = st0['H2']
for z in z_plot:
    N  = max(np.log(1/(1+z)), N_start)
    st = get_state(sol_opt, N, f_opt)
    w_ev.append(st['w_phi'])
    Op_ev.append(st['Omega_phi'])
    phi_ev.append(st['phi'])
    E_ev.append(np.sqrt(st['H2']/H2_0))

fig, axes = plt.subplots(2,2,figsize=(13,9))
fig.suptitle(
    f'TIFA BEST-FIT: f={f_opt:.3f} M_Pl | '
    f'w(0)={w0_opt:.3f} | '
    f'H0={H0_opt:.2f} | '
    f'χ²={c2_opt:.2f} vs ΛCDM χ²=15.81',
    fontsize=11
)

# w(z)
ax = axes[0,0]
ax.plot(z_plot, w_ev, 'b-', lw=2.5,
        label=f'TIFA w(0)={w0_opt:.3f}')
ax.axhline(-1.0, color='k', ls=':', lw=1.5,
           label='ΛCDM w=-1')
ax.fill_between(z_plot,
    [-0.95]*len(z_plot),
    [-0.65]*len(z_plot),
    alpha=0.15, color='green',
    label='DESI DR1 preferred')
ax.set_xlabel('z'); ax.set_ylabel('w(z)')
ax.set_title('Equation of State w(z)')
ax.set_xlim([0,3]); ax.grid(alpha=0.3)
ax.legend(fontsize=9)

# Omega_phi(z)
ax = axes[0,1]
ax.plot(z_plot, Op_ev, 'r-', lw=2.5,
        label='Ω_φ(z)')
ax.axhline(Omega_DE, color='k', ls=':',
           lw=1.5, label=f'Ω_DE={Omega_DE}')
ax.set_xlabel('z'); ax.set_ylabel('Ω_φ')
ax.set_title('Dark Energy Fraction')
ax.set_xlim([0,3]); ax.grid(alpha=0.3)
ax.legend(fontsize=9)

# phi(z)
ax = axes[1,0]
ax.plot(z_plot, phi_ev, 'm-', lw=2.5,
        label='φ(z)')
ax.axhline(np.pi*f_opt, color='r', ls=':',
           lw=1.5,
           label=f'πf={np.pi*f_opt:.3f}')
ax.axhline(phi_i_opt, color='g', ls='--',
           lw=1.5,
           label=f'φ_i={phi_i_opt:.3f}')
ax.set_xlabel('z'); ax.set_ylabel('φ [M_Pl]')
ax.set_title('Field Evolution φ(z)')
ax.set_xlim([0,3]); ax.grid(alpha=0.3)
ax.legend(fontsize=9)

# H(z)/H0
ax = axes[1,1]
ax.plot(z_plot, E_ev, 'b-', lw=2.5,
        label='TIFA')
E_LCDM = [np.sqrt(0.315*(1+z)**3+0.685)
           for z in z_plot]
ax.plot(z_plot, E_LCDM, 'k:', lw=1.5,
        label='ΛCDM')
ax.set_xlabel('z'); ax.set_ylabel('H(z)/H₀')
ax.set_title('Hubble Rate')
ax.set_xlim([0,3]); ax.grid(alpha=0.3)
ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig('tifa_best_fit.png',
            dpi=150, bbox_inches='tight')
print("  Saved: tifa_best_fit.png")
print()
print("="*65)
print("SUMMARY")
print("="*65)
print(f"""
  TIFA BEST FIT (Ω_φ=0.685 constrained):
  ────────────────────────────────────────
  f         = {f_opt:.4f} M_Pl
  φ_i       = {phi_i_opt:.6f} M_Pl
  w(z=0)    = {w0_opt:.4f}
  H0        = {H0_opt:.2f} km/s/Mpc
  χ²        = {c2_opt:.3f}
  χ²_ΛCDM   = 15.805

  Δχ²  = {c2_opt-15.805:+.3f}
  ΔAIC = {c2_opt-15.805-2:+.3f}  (TIFA has 1 param vs 2)
  ΔBIC = {c2_opt-15.805-np.log(N_data):+.3f}

  STATUS: TIFA {'COMPETITIVE WITH' if c2_opt < 20
  else 'BEHIND'} ΛCDM
""")
```
